##**Brief Description of Game of Thrones**








*Game of Thrones* is an American fantasy drama television series created by David Benioff and D. B. Weiss for HBO. It is an adaptation of A Song of Ice and Fire, a series of fantasy novels by George R. R. Martin, the first of which is A Game of Thrones. The show premiered on HBO in the United States on **April 17, 2011**, and concluded on **May 19, 2019**, with 73 episodes broadcast over eight seasons.

Set on the fictional continents of Westeros and Essos, Game of Thrones has a large ensemble cast and follows several story arcs throughout the course of the show. The first major arc concerns the Iron Throne of the Seven Kingdoms of Westeros through a web of political conflicts among the noble families either vying to claim the throne or fighting for independence from whoever sits on it. The second major arc focuses on the last descendant of the realm's deposed ruling dynasty, who has been exiled to Essos and is plotting to return and reclaim the throne. The third follows the Night's Watch, a military order defending the realm against threats from beyond Westeros' northern border.

Game of Thrones attracted a record viewership on HBO and has a broad, active, and international fan base. Critics have praised the series for its acting, complex characters, story, scope, and production values, although its frequent use of nudity and violence (including sexual violence) generated controversy. The final season received significant criticism for its reduced length and creative decisions, with many considering it a disappointing conclusion. The series received 59 Primetime Emmy Awards, the most by a drama series, including Outstanding Drama Series in 2015, 2016, 2018 and 2019. Its other awards and nominations include three Hugo Awards for Best Dramatic Presentation, a Peabody Award, and five nominations for the Golden Globe Award for Best Television Series  Drama.

**A prequel series, House of the Dragon, premiered on HBO in 2022.**

```python
import matplotlib.pyplot as plt

data = {
    'Season': [1, 2, 3, 4, 5, 6, 7, 8],
    'Episodes': [10, 10, 10, 10, 10, 10, 7, 6],
    'First Aired': ['April 17, 2011', 'April 1, 2012', 'March 31, 2013', 'April 6, 2014',
                    'April 12, 2015', 'April 24, 2016', 'July 16, 2017', 'April 14, 2019'],
    'Last Aired': ['June 19, 2011', 'June 3, 2012', 'June 9, 2013', 'June 15, 2014',
                   'June 14, 2015', 'June 26, 2016', 'August 27, 2017', 'May 19, 2019'],
    'Avg. U.S. Viewers (millions)': [2.52, 3.80, 4.97, 6.84, 6.88, 7.69, 10.26, 11.99]
}

# Create a DataFrame
got_df = pd.DataFrame(data)

# Convert the 'First Aired' and 'Last Aired' to datetime
got_df['First Aired'] = pd.to_datetime(got_df['First Aired'])
got_df['Last Aired'] = pd.to_datetime(got_df['Last Aired'])

# Display the DataFrame
got_df
```

```python
# Summary of basic statistics
summary_statistics = got_df['Avg. U.S. Viewers (millions)'].describe()

summary_statistics
```

```python
# Calculate the viewership change between seasons and add it to the DataFrame
got_df['Viewership Change (millions)'] = got_df['Avg. U.S. Viewers (millions)'].diff()

# Plot the change in viewership from season to season
plt.figure(figsize=(10, 6))
plt.bar(got_df['Season'], got_df['Viewership Change (millions)'])
plt.title('Change in Average U.S. Viewership Between Seasons')
plt.xlabel('Season')
plt.ylabel('Change in Viewership (millions)')
plt.grid(True)
plt.show()

# Correcting the inline reference to calculate the viewership change between seasons 3 and 5
season_3_to_5_increase = got_df.loc[got_df['Season'] == 5, 'Avg. U.S. Viewers (millions)'].values[0] - \
                         got_df.loc[got_df['Season'] == 3, 'Avg. U.S. Viewers (millions)'].values[0]

description = (f"The viewership increased by {season_3_to_5_increase:.2f} million between "
               "Seasons 3 and 5.")

# Rerun the summary statistics as well
summary_statistics = got_df['Avg. U.S. Viewers (millions)'].describe()

summary_statistics, description
```

The bar graph shows the change in average U.S. viewership between each season of "Game of Thrones". The bars represent the increase or decrease in the number of viewers (in millions) compared to the previous season. Here's a simple interpretation:



*   Between Season 1 and Season 2, the viewership increased by about 1.28 million.
*   The increase continued from Season 2 to Season 3, with an approximate growth of 1.17 million.
*   From Season 3 to Season 4, there was a larger increase in viewership of about 1.87 million.
*   The change from Season 4 to Season 5 was minimal, showing a slight increase in viewership.
*   Season 5 to Season 6 again saw a modest increase in viewership.
*   There was a substantial jump in viewership from Season 6 to Season 7, suggesting heightened interest as the series neared its conclusion.
*   The viewership from Season 7 to Season 8, the final season, increased significantly, indicating peak interest in the series' outcome.
*   The general trend is that the show gained viewers with each season, with particularly notable spikes in the later seasons, reflecting growing popularity and audience engagement as the series progressed.

## Viewership for each season of "Game of Thrones."

```python
import matplotlib.pyplot as plt
import numpy as np

viewership = np.array([2.52, 3.80, 4.97, 6.84, 6.88, 7.69, 10.26, 11.99])

# Season numbers
seasons = np.arange(1, len(viewership) + 1)

# Plotting the viewership as a colorful bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(seasons, viewership, color=plt.cm.Spectral(np.linspace(0, 1, len(viewership))))

# Adding a title and labels
plt.title('Game of Thrones Viewership by Season', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Average Viewership (millions)', fontsize=12)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', ha='center', fontsize=10)

plt.show()

```
